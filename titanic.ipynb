```python
titanic_df=titanic_df.drop(['PassengerId','SibSp'],axis=1)
```

```python
print("======Data Headers=======")
titanic_df.head()
```

```python
titanic_df["Embarked"]=titanic_df["Embarked"].fillna("S")
```

```python
print("======Data Headers=======")
titanic_df.head()
```

```python
titanic_df['Pclass']=titanic_df['Pclass'].map({
    1:'Luxuary class',
    2:'Economy class',
    3:'Lower class'
})
```

```python
print("titanic_df.head()")
```

```python
titanic_df.head()
```

```python
titanic_df["Embarked"]=titanic_df["Embarked"].map({
    "S":"Sydney",
    "C":"chicago",
    "Q":"Queensland"
})
```

```python
titanic_df.head()
```

```python
titanic_df ['Embarked'] = titanic_df ['Embarked'].map({
    'C':'Chicago',
    'Q':'Quensland',
    'S':'Sydney'
})
```

```python
titanic_df.head()
```

```python
titanic_df['Survived'] = titanic_df['Survived'].map({
    0:'Died',
    1:'Survived'
})
```

```python
ax = sns.countplot(x = 'Pclass', hue = 'Survived', palette = 'Set1', data=titanic_df)
ax.set(title = 'Passenger status(Survived/died) against Pclass', xlabel = 'Pclass', ylabel = 'Total')
plt.show()
```

```python
print(pd.crosstab(titanic_df["Sex"],titanic_df.Survived))
ax = sns.countplot(x = 'Sex', hue = 'Survived', palette = 'Set2' , data=titanic_df)
ax.set(title = 'Total survivals according to Sex', xlabel = 'Sex', ylabel='Total')
plt.show()
```

```python
interval = (0,18,25,50,100)
categories = ['Children', 'Teens', 'Adults','Old']
titanic_df['Age_cats'] = pd.cut(titanic_df.Age, interval,labels = categories)
ax = sns.countplot(x = 'Age_cats', data=titanic_df, hue='Survived',palette='Set3')
ax.set(xlabel='Age Categorical',ylabel='Total',title='Age Categorical Survival Distribution')
plt.show()
```
